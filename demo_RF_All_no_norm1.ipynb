```python
import numpy as np
import pandas as pd 
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score, accuracy_score
```

```python
pd.set_option('max_columns', 500)
pd.set_option('max_colwidth', -1)
pd.set_option('max_rows', 500)
```

```python
inFile = '/home/rohit/SRH/CaseStudy1/dataDemo/dataFiles/All_Features_No_Norm_11kURLs.csv'
```

```python
dfData = pd.read_csv(inFile,sep=',', index_col=0)
```

```python
dfData.head(2)
```

```python
dfData.shape
```

```python
dfData = dfData.drop(['domain', 'url'], axis = 1)
```

```python
dfData.head(2)
```

```python
# y is the urlType column
# X is the feature columns
```

```python
X = dfData.iloc[:, 1:].values
y = dfData.iloc[:, 0:1].values
```

```python
X.shape
```

```python
y.shape
```

```python
## We want to create approx 1000 data points as a Validation set.
## Overall data is around 11k data points, so splitting first with Train:Test = 90:10
```

```python
X_remaining, X_validation, y_remaining, y_validation = train_test_split( X, y, test_size = 0.1, random_state = 100)
```

```python
X_validation.shape
```

```python
y_validation.shape
```

```python
## Now the remaining data split in 80:20 ratio, as we did for the training of the models
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_remaining, y_remaining, test_size = 0.2, random_state = 100)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Random Forest
```

```python
model_tree = RandomForestClassifier(random_state=100, n_estimators=100)
model_tree.fit(X_train, y_train)
#print(model_tree.feature_importances_)
#sel_model_tree = sklearn.feature_selection.SelectFromModel(estimator=model_tree, prefit=True, threshold='mean')  
#X_train_sfm_tree = sel_model_tree.transform(X_train)
#print(sel_model_tree.get_support())
```

```python
rf_predictions = model_tree.predict(X_test)
```

```python
print (f"Train Accuracy :: {accuracy_score(y_train, model_tree.predict(X_train))}")
print (f"Test  Accuracy :: {accuracy_score(y_test, rf_predictions)}")
```

```python
rf_probs = model_tree.predict_proba(X_test)[:, 1]
```

```python
# Calculate roc auc/model accuracy for test
roc_value = roc_auc_score(y_test, rf_probs)
```

```python
roc_value
```

```python
rf_predictions_val = model_tree.predict(X_validation)
```

```python
print (f"Validation  Accuracy :: {accuracy_score(y_validation, rf_predictions_val)}")
```

```python
rf_probs_val = model_tree.predict_proba(X_validation)[:, 1]
```

```python
# Calculate roc auc/model accuracy for validation
roc_value_val = roc_auc_score(y_validation, rf_probs_val)
```
